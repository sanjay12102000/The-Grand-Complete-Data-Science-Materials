```python
import numpy as np

def parse_pdb(pdb_file_path):
    pdb_data = {'chains': []}

    with open(pdb_file_path, 'r') as pdb_file:
        current_chain = None
        current_residue = None

        for line in pdb_file:
            record_type = line[:6].strip()
            if record_type == 'ATOM':
                atom_info = {
                    'atom_name': line[12:16].strip(),
                    'amino_acid_name': line[17:20].strip(),
                    'chain_id': line[21],
                    'residue_number': int(line[22:26]),
                    'atom_position': np.array([float(line[30:38]), float(line[38:46]), float(line[46:54])])
                }

                if current_chain is None or atom_info['chain_id'] != current_chain['chain_id']:
                    current_chain = {'chain_id': atom_info['chain_id'], 'residues': []}
                    pdb_data['chains'].append(current_chain)

                if current_residue is None or atom_info['residue_number'] != current_residue['residue_number']:
                    current_residue = {'residue_number': atom_info['residue_number'], 'amino_acid_name': atom_info['amino_acid_name'], 'atoms': []}
                    current_chain['residues'].append(current_residue)

                current_residue['atoms'].append(atom_info)

    return pdb_data

def calculate_euclidean_distance(coord1, coord2):
    return np.linalg.norm(coord1 - coord2)

def find_contacting_atoms(pdb_data, amino_acid_residue_number, distance_cutoff=5.0):
    amino_acid_chain_id = 'A'  # Change this to the appropriate chain ID
    amino_acid_residue_number = int(amino_acid_residue_number)  # Ensure it's an integer

    amino_acid_atoms = None
    amino_acid_residue = None

    for chain in pdb_data['chains']:
        if chain['chain_id'] == amino_acid_chain_id:
            for residue in chain['residues']:
                if residue['residue_number'] == amino_acid_residue_number and residue['amino_acid_name'] not in ['HOH', 'UNK']:
                    amino_acid_residue = residue
                    amino_acid_atoms = residue['atoms']
                    break
            if amino_acid_residue:
                break

    if amino_acid_residue is not None:
        contacting_atoms = []

        for chain in pdb_data['chains']:
            for residue in chain['residues']:
                if residue != amino_acid_residue and residue['amino_acid_name'] not in ['HOH', 'UNK']:
                    for atom_info in amino_acid_atoms:
                        amino_acid_coords = atom_info['atom_position']
                        for contacting_atom_info in residue['atoms']:
                            residue_coords = contacting_atom_info['atom_position']
                            distance = calculate_euclidean_distance(amino_acid_coords, residue_coords)

                            if distance < distance_cutoff:
                                contacting_atom_info['distance'] = distance
                                contacting_atoms.append(contacting_atom_info)

        return contacting_atoms

    else:
        print(f"Amino acid with residue number {amino_acid_residue_number} not found.")
        return []

# Example usage with multiple PDB files and amino acid numbers:
pdb_file_paths = directory
amino_acid_numbers_of_interest = Resid  # Change this to the specific amino acids you are interested in

for pdb_file_path, amino_acid_number_of_interest in zip(pdb_file_paths, amino_acid_numbers_of_interest):
    pdb_data = parse_pdb(pdb_file_path)
    contacting_atoms = find_contacting_atoms(pdb_data, amino_acid_number_of_interest)

    print(f"Results for {pdb_file_path}, Amino Acid {amino_acid_number_of_interest}:")
    for atom_info in contacting_atoms:
        print(f"Contacting Residue: {atom_info['amino_acid_name']} {atom_info['residue_number']} in Chain {atom_info['chain_id']}")
        print(f"Atom Name: {atom_info['atom_name']}, Atom Position: {atom_info['atom_position']}, Distance: {atom_info['distance']}")
        print()
```
